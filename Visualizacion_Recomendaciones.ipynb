```
import pandas as pd
import numpy as np
import json

# reading the JSON data using json.load()
file = 'Recomendaciones_black.txt'
with open(file) as data_file:
    dict_data = json.load(data_file)

# converting json dataset from dictionary to dataframe
train = pd.DataFrame.from_dict(dict_data, orient='index')
train.reset_index(level=0, inplace=True)


train = train.rename(columns={'index': 'CLI_ID'})
train = train.iloc[0:110]

train.replace('None', np.nan, inplace=True)

train

#train_aux = train.fillna(value=pd.np.nan, inplace=True)

#train_aux

```
